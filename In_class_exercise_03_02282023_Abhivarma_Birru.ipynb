Question 2 (10 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction. 

```python
# You code here (Please add comments in the code):
from six.moves import urllib
import pandas as pd 
txtList = []
urlForText = "https://raw.githubusercontent.com/Abhivarma-Birru-Unt/Abhivarma_INFO5731_Spring2023/main/sampledata.txt"
txtFile = urllib.request.urlopen(urlForText)  #This function is used to decode the url and get the data from it it is a predefined function
for i in txtFile:
    decoded_data = i.decode("utf-8").replace('\r\n', '')    #this statement is used to decode the data present in the url and convert it into standard format
    if decoded_data: 
        txtList.append(decoded_data)       #appending the decoded data to a list
txtDataFrame = pd.DataFrame (txtList, columns = ['Text'])   #converting the list to data frame using pamdas
print(txtDataFrame)
```

```python
#To find Number of Sentences in a data frame

txtDataFrame['Number of sentences'] = txtDataFrame['Text'].apply(lambda y: len(str(y).split(".")))
txtDataFrame
```

```python
#To find Number of words in a data frame

txtDataFrame['Number of Words'] = txtDataFrame['Text'].apply(lambda y: len(str(y).split(" ")))
txtDataFrame
```

```python
#To find Number of characters in data frame

txtDataFrame['Number of Characters'] = txtDataFrame['Text'].str.len()
txtDataFrame
```

```python
#To find the average word length of the list


def avgWordLength(x):
    wordList = x.split() #here the sentence is being splitted into parts and being assigned to variable wordList
    if len(wordList) != 0:  #here we are checking if the variable is empty or not
        return(sum(len(deep) for deep in wordList)/len(wordList))
    else:
        return None

txtDataFrame['Avg Word Length'] = txtDataFrame['Text'].apply(lambda z: avgWordLength(z))
txtDataFrame
```

```python
#To find the number of special characters present in the list

def specialCharacters(x):
    read = 0
    for i in range(len(x)):
        if not((x[i].isalpha()) and (x[i].isdigit())):
            read = read + 1
    return read

txtDataFrame['Number of special characters'] = txtDataFrame['Text'].apply(lambda z: specialCharacters(z))
txtDataFrame
```

```python
#To find the number of numerics present in the list

txtDataFrame['Number of number'] = txtDataFrame['Text'].apply(lambda x: len([x for x in x.split() if x.isdigit()])) #here we are considering it to be a seperate numeric if it is seperated with a space using the split function
txtDataFrame
```

```python
#To find number of uppercase words in a list

txtDataFrame['Number of upper case words'] = txtDataFrame['Text'].apply(lambda x: len([x for x in x.split() if x.isupper()])) #using split function we are finding the words with completely upper case
txtDataFrame
```
